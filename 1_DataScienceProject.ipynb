```python
start=dt.date(2006,1,1)
end=dt.date(2016,1,1)
print(start)
print(end)
```

```python
stocklist=['BAC','C','GS','JPM','MS','WFC']
```

```python
df=pdr.DataReader(stocklist,'yahoo',start,end)
df.columns.names = ['Stock Info','Bank Ticker']
df.head()
```

```python
df.index
```

```python
close=df.Close.max()
close
#df.xs(key='Close',axis=1,level='Stock Info').max()  #xs stands for cross section
```
